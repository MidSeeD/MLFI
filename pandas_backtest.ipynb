## Pandas for time series

**Dữ liệu ngày và giờ có nhiều loại khác nhau mà chúng ta sẽ thảo luận ở đây:**

- *Time stamps* tham chiếu các thời điểm cụ thể trong thời gian (ví dụ: ngày 4 tháng 7 năm 2015 lúc 7:00 sáng).
- *Time intervals* and *periods* tham chiếu khoảng thời gian giữa điểm bắt đầu và điểm kết thúc cụ thể; ví dụ: năm 2015. Các khoảng thời gian thường đề cập đến một trường hợp đặc biệt của các khoảng thời gian trong đó mỗi khoảng thời gian có độ dài đồng đều và không chồng lên nhau (ví dụ: các khoảng thời gian dài 24 giờ bao gồm các ngày).
- *Time deltas* hoặc *durations* tham chiếu khoảng thời gian chính xác (ví dụ: khoảng thời gian 22,56 giây).

### Resampling and converting frequencies

One common need for time series data is resampling at a higher or lower frequency.
This can be done using the ``resample()`` method, or the much simpler ``asfreq()`` method.
The primary difference between the two is that ``resample()`` is fundamentally a *data aggregation*, while ``asfreq()`` is fundamentally a *data selection*.

Taking a look at the Google closing price, let's compare what the two return when we down-sample the data.
Here we will resample the data at the end of business year:

| Code   | Description         | Code   | Description          |
|--------|---------------------|--------|----------------------|
| ``D``  | Calendar day        | ``B``  | Business day         |
| ``W``  | Weekly              |        |                      |
| ``M``  | Month end           | ``BM`` | Business month end   |
| ``Q``  | Quarter end         | ``BQ`` | Business quarter end |
| ``A``  | Year end            | ``BA`` | Business year end    |
| ``H``  | Hours               | ``BH`` | Business hours       |
| ``T``  | Minutes             |        |                      |
| ``S``  | Seconds             |        |                      |
| ``L``  | Milliseonds         |        |                      |
| ``U``  | Microseconds        |        |                      |
| ``N``  | nanoseconds         |        |                      |

The monthly, quarterly, and annual frequencies are all marked at the end of the specified period.
By adding an ``S`` suffix to any of these, they instead will be marked at the beginning:

| Code    | Description            || Code    | Description            |
|---------|------------------------||---------|------------------------|
| ``MS``  | Month start            ||``BMS``  | Business month start   |
| ``QS``  | Quarter start          ||``BQS``  | Business quarter start |
| ``AS``  | Year start             ||``BAS``  | Business year start    |

Additionally, you can change the month used to mark any quarterly or annual code by adding a three-letter month code as a suffix:

- ``Q-JAN``, ``BQ-FEB``, ``QS-MAR``, ``BQS-APR``, etc.
- ``A-JAN``, ``BA-FEB``, ``AS-MAR``, ``BAS-APR``, etc.

In the same way, the split-point of the weekly frequency can be modified by adding a three-letter weekday code:

- ``W-SUN``, ``W-MON``, ``W-TUE``, ``W-WED``, etc.

On top of this, codes can be combined with numbers to specify other frequencies.
For example, for a frequency of 2 hours 30 minutes, we can combine the hour (``H``) and minute (``T``) codes as follows:

```python
#asfreq
stock.asfreq('BA')
```

```python
#resample
stock.resample('BA').mean()
```

```python
# custom agg fucntion 
def custom_mean(container):
  return np.mean(container)
```

```python
stock.resample('BA').agg(custom_mean)
```

```python
stock.plot(alpha=0.5, style='-')
stock.resample('BQ').mean().plot(style='y-')
stock.asfreq('BQ').plot(style='--');
plt.legend(['input', 'resample', 'asfreq'],
           loc='upper left');
```

Notice the difference: at each point, ``resample`` reports the *average of the previous year*, while ``asfreq`` reports the *value at the end of the year*.

#### Define strategy 
image.png

```python
# define strategy 
def rsi_strategy(rsi):
  if rsi > 70:
    return -1
  if rsi < 30:
    return 1
  return 0
df['strategy'] = df['rsi_window_14'].apply(rsi_strategy).shift(1)
df.tail()
```

```python
# backtest
total_cash = 100_000_000
my_trades = []
trade_format = {
    'size': None,
    'entryTime': None,
    'entryPrice': None,
    'exitTime': None,
    'exitPrice': None
}
for trading_day in df.index:
  trading_data = df.loc[:trading_day]
  strategy = trading_data.strategy[-1]
  stock_open_price = trading_data.Open[-1]
  stock_close_price = trading_data.Close[-1]

  #compute equity
  equity = 0
  equity += total_cash
  for trade in my_trades:
    if trade['exitTime'] is None:
      equity += trade['size'] * stock_close_price

  #trade
  if strategy == 1: #buy
    if total_cash >= stock_open_price:
      size = int(total_cash / stock_open_price)
      my_trades.append({
          'size': size,
          'entryTime': trading_day,
          'entryPrice': stock_open_price,
          'exitTime': None,
          'exitPrice': None
      })
      total_cash -= size * stock_open_price
  elif strategy == -1 or trading_day == df.index[-1]:  #close
    for trade in my_trades :
      if trade['exitTime'] is None:
        if (trading_day - trade['entryTime']).days > 2:
          trade['exitTime'] = trading_day
          trade['exitPrice'] = stock_open_price
          total_cash += trade['size'] * stock_open_price

  #update df
  df.loc[trading_day, 'equity'] = equity
```

```python
my_trade_df['PnL'] = my_trade_df['size'] * (my_trade_df['exitPrice'] - my_trade_df['entryPrice'])
my_trade_df['duration'] = my_trade_df['exitTime'] - my_trade_df['entryTime']
my_trade_df
```

### With backtesting library

```python
from backtesting import Backtest
from backtesting import Strategy
from backtesting.lib import crossover
import math

class RsiOscillator(Strategy):

    upper_bound = 70
    lower_bound = 30
    rsi_window = 14
    def cash(self):
      return self.equity - sum([trade.value for trade in self.trades])

    # Do as much initial computation as possible
    def init(self):
      self.rsi = self.I(rsi, pd.Series(self.data.Close), self.rsi_window)

    def next(self):
      cash = self.cash()
      if self.rsi[-1] > self.upper_bound:
          if len(self.trades) != 0:
            for trade in self.trades:
              if (self.data.index[-1] - trade.entry_time).days > 2:
                trade.close()
      elif self.lower_bound > self.rsi[-1]:
            self.buy()


bt = Backtest(df, RsiOscillator, cash=100_000_000, commission=0)
stats = bt.run()


# some notes:
# _____BUY FUNCTION___
# def buy(self, *, size=.9999, limit=None, stop=None, sl=None, tp=None)
# Place a new long order. For explanation of parameters, see Order and its properties.

# _____SELL FUNCTION___
# def sell(self, *, size=.9999, limit=None, stop=None, sl=None, tp=None)
# Place a new short order. For explanation of parameters, see Order and its properties.

# _____CLOSE FUNCTION___
# def close(self, portion=1.0)
# Place new Order to close portion of the trade at next market price.
```
